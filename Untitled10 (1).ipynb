```python
columns=['Price','MRP']

for i in columns:
        # Assuming df is your DataFrame containing the dataset
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1 * IQR
        upper_bound = Q3 + 1 * IQR


        # Alternatively, replace outliers with the nearest non-outlier value
        df[i] = df[i].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

        
```

# Market Leader in Each Category

```python
category_leader = df.groupby(['Category', 'Brand'])['Units Sold'].sum().reset_index()
market_leader = category_leader.groupby('Category').apply(lambda x: x.loc[x['Units Sold'].idxmax()]).reset_index(drop=True)
market_leader
```

# Brand Market Share Based on Units Sold and Revenue

```python
brand_totals = df.groupby('Brand').agg({'Units Sold': 'sum', 'Price': 'sum'})
brand_totals['Market Share (Units Sold)'] = (brand_totals['Units Sold'] / df['Units Sold'].sum())
brand_totals['Market Share (Revenue)'] = (brand_totals['Price'] * brand_totals['Units Sold']) / (df['Price'] * df['Units Sold']).sum()
```

```python
brand_totals['Market Share (Units Sold)'].nlargest(5)
```

```python
brand_totals['Market Share (Revenue)'].nlargest(5)
```

# Brand Level RPI (Relative Pricing Index)

```python
# Calculate average price for each brand
brand_avg_price = df.groupby('Brand')['Price'].mean()

# Calculate RPI (Relative Pricing Index)
avg_price_all_brands = df['Price'].mean()
brand_rpi = brand_avg_price / avg_price_all_brands

# Display brand RPI
print(brand_rpi.nlargest(5))
```

# Brand Discounting Trend

```python
df['Discount'] = ((df['MRP'] - df['Price']) / df['MRP']) * 100
df['Discount'] = df['Discount'].abs()
```

```python
top_brands = df.groupby('Brand')['Units Sold'].sum().nlargest(5).index
```

```python
brand_discount_daily = df[df['Brand'].isin(top_brands)].groupby(['Brand', 'Date'])['Discount'].mean().reset_index()
plt.figure(figsize=(12, 6))
for brand in top_brands:
    data = brand_discount_daily[brand_discount_daily['Brand'] == brand]
    plt.plot(data['Date'], data['Discount'], label=brand)

plt.title('Brand Daily Discounting Trend')
plt.xlabel('Date')
plt.ylabel('Discount')
plt.legend()
plt.show()
```

```python
brand_discount_weekly = df.groupby(['Brand', 'Week'])['Discount'].mean().reset_index()
plt.figure(figsize=(12, 6))
for brand in top_brands:
    data = brand_discount_weekly[brand_discount_weekly['Brand'] == brand]
    plt.plot(data['Week'], data['Discount'], label=brand)

plt.title('Brand Weekly Discounting Trend')
plt.xlabel('Week')
plt.ylabel('Discount')
plt.legend()
plt.show()
```

# Brand Average Price Trend

```python
# Daily average price trend for top brands by units sold
brand_avg_price_daily = df[df['Brand'].isin(top_brands)].groupby(['Brand', 'Date'])['Price'].mean().reset_index()

# Plot daily average price trend for top brands
plt.figure(figsize=(12, 6))
for brand in top_brands:
    data = brand_avg_price_daily[brand_avg_price_daily['Brand'] == brand]
    plt.plot(data['Date'], data['Price'], label=brand)

plt.title('Brand Daily Average Price Trend')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.show()

```

```python
# Weekly average price trend for top brands by units sold
brand_avg_price_weekly = df.groupby(['Brand', 'Week'])['Price'].mean().reset_index()

# Plot weekly average price trend for top brands
plt.figure(figsize=(12, 6))
for brand in top_brands:
    data = brand_avg_price_weekly[brand_avg_price_weekly['Brand'] == brand]
    plt.plot(data['Week'], data['Price'], label=brand)

plt.title('Brand Weekly Average Price Trend')
plt.xlabel('Week')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

# category growth trend based on units sold

```python
# Daily category growth trend based on units sold
category_growth_daily = df.groupby(['Category', 'Date'])['Units Sold'].sum().reset_index()

# Plot daily category growth trend
plt.figure(figsize=(12, 6))
for category in df['Category'].unique():
    data = category_growth_daily[category_growth_daily['Category'] == category]
    plt.plot(data['Date'], data['Units Sold'], label=category)

plt.title('Category Daily Growth Trend')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

```python
# Weekly category growth trend based on units sold
category_growth_weekly = df.groupby(['Category', 'Week'])['Units Sold'].sum().reset_index()

# Plot weekly category growth trend
plt.figure(figsize=(12, 6))
for category in df['Category'].unique():
    data = category_growth_weekly[category_growth_weekly['Category'] == category]
    plt.plot(data['Week'], data['Units Sold'], label=category)

plt.title('Category Weekly Growth Trend')
plt.xlabel('Week')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

# products in each category based on units sold

```python
# Identify top products in each category based on units sold
top_products = df.groupby(['Category', 'Prod ID'])['Units Sold'].sum().reset_index()
top_products = top_products.sort_values(['Category', 'Units Sold'], ascending=[True, False]).groupby('Category').head(1)

plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Units Sold', hue='Prod ID', data=top_products)
plt.title('Top Products in Each Category based on Units Sold')
plt.xlabel('Category')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

# market share by category

```python
# Calculate market share by category
category_market_share = df.groupby('Category')['Units Sold'].sum().reset_index()

# Plotting a pie chart for market share by category
plt.figure(figsize=(8, 8))
plt.pie(category_market_share['Units Sold'], labels=category_market_share['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Market Share by Category (Units Sold)')
plt.show()
```

# brand market share

```python
brand_market_share = df.groupby('Brand')['Units Sold'].sum().sort_values(ascending=False).reset_index()

# Plotting a bar chart for brand market share
plt.figure(figsize=(12, 6))
sns.barplot(x='Brand', y='Units Sold', data=brand_market_share.head(5))
plt.title('Top 5 Brands by Market Share (Units Sold)')
plt.xlabel('Brand')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```
